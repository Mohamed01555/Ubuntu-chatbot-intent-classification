```python
import numpy as np
import pandas as pd
```

```python
import json
with open('data.json','r') as f:
    dataset = json.loads(f.read())
```

```python
dataset = {dataset['sentences'][i]['text'] : dataset['sentences'][i]['intent'] for i in range(len(dataset['sentences']))}
```

```python
dataset
```

```python
X = list(dataset.keys())
y = list(dataset.values())
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X ,y ,test_size=0.2) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english',ngram_range=(1,3))
x_train = tfidf.fit_transform(x_train)
x_test = tfidf.transform(x_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(class_weight='balanced')
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print(classification_report(y_test,y_pred))
pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
def test(input_text):
    input_text = tfidf.transform([input_text])
    return f'This text belongs to {rf.predict(input_text)[0]} with probability {max(rf.predict_proba(input_text)[0])}'
```

```python
test('Upgrading From 11.04 (32bit) to 12.04 (64bit)')
```
